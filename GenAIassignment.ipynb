```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


url = "https://raw.githubusercontent.com/HarshitEmpire/GenAIassignment/main/employee_productivity.csv"
df = pd.read_csv(url)


print("Dataset Preview:")
print(df.head())


X = df[['Experience','Training','WorkingHours','Projects']]
y = df['Productivity']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LinearRegression()
model.fit(X_train, y_train)


print("\nRegression Coefficients:")
print("Experience:", model.coef_[0])
print("Training:", model.coef_[1])
print("WorkingHours:", model.coef_[2])
print("Projects:", model.coef_[3])
print("Intercept:", model.intercept_)


y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nModel Performance:")
print("Mean Squared Error:", mse)
print("R2 Score:", r2)

print("\nPredicted vs Actual:")
print(pd.DataFrame({'Predicted': y_pred, 'Actual': y_test.values}))
```
