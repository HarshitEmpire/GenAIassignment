```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Sample dataset
area = np.array([800, 900, 1000, 1200, 1500, 1800])
bedrooms = np.array([2, 2, 3, 3, 4, 4])
price = np.array([200, 220, 250, 300, 380, 450])


X = np.column_stack((area, bedrooms))
y = price


model = LinearRegression()
model.fit(X, y)


area_range = np.linspace(area.min(), area.max(), 20)
bedroom_range = np.linspace(bedrooms.min(), bedrooms.max(), 20)
area_grid, bedroom_grid = np.meshgrid(area_range, bedroom_range)


price_pred = model.predict(
    np.column_stack((area_grid.ravel(), bedroom_grid.ravel()))
)
price_pred = price_pred.reshape(area_grid.shape)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(area, bedrooms, price)


ax.plot_surface(area_grid, bedroom_grid, price_pred, alpha=0.5)

# Labels
ax.set_xlabel("Area (sq ft)")
ax.set_ylabel("Number of Bedrooms")
ax.set_zlabel("House Price")
ax.set_title("3D Visualization of House Price Regression Model")

plt.show()
```
